```python
import pandas as pd
import re

# Simulação da sua lista de dados
dados_brutos = precos_carrefour[0]

# Dicionário de estados → regiões do Brasil
regioes_brasil = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste',
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

dados_processados = []
avisos = []

for linha in dados_brutos:
    try:
        # 1. Remove preços por unidade (ex: R$ 185,90/kg)
        linha_limpa = re.sub(r"R\$ ?\d+,\d{2}/(kg|g|l|ml|un|100ml|litro)", "", linha, flags=re.IGNORECASE).strip()

        # 2. Separa por quebra de linha
        partes = linha_limpa.split("\n")

        # 3. Extrai o primeiro preço que aparece
        preco_bruto = next((p for p in partes if "R$" in p), None)
        preco = float(re.search(r"R\$ ?(\d+,\d{2})", preco_bruto).group(1).replace(",", ".")) if preco_bruto else None

        # 4. Nome do produto
        produto = partes[0].strip()

        # 5. Extrai loja, cidade e estado
        endereco_match = re.search(r"(.*?)(R\$ ?\d+,\d{2})?\s*(.+), (.+?) - (\w{2})$", linha_limpa)
        if endereco_match:
            loja_com_preco = endereco_match.group(3).strip()
            loja = re.sub(r"^R\$ ?\d+,\d{2}", "", loja_com_preco).strip()
            cidade = endereco_match.group(4).strip()
            estado = endereco_match.group(5).strip()
        else:
            raise ValueError("Endereço não encontrado")

        # 6. Define a região com base no estado
        regiao = regioes_brasil.get(estado, "Desconhecida")

        # 7. Monta o dicionário de dados
        dados_processados.append({
            "produto": produto,
            "preco": preco,
            "loja": loja,
            "cidade": cidade,
            "estado": estado,
            "regiao": regiao
        })

    except Exception as e:
        avisos.append(f"Erro ao processar linha: {linha}\n-> {e}")

# Cria o DataFrame final
df_tilapia = pd.DataFrame(dados_processados)

# Exibe os avisos, se houver
if avisos:
    for aviso in avisos:
        print(aviso)
else:
    print("✅ Todos os dados foram processados corretamente.")
```

```python
precos_carrefour1 = precos_carrefour1.rename(columns={'cidade': 'cidade2', 'cidade_estado': 'cidade'})
precos_carrefour1
```

```python
# Extrair o primeiro preço encontrado (R$ xx,xx)
precos_carrefour1['preco'] = precos_carrefour1['preco'].str.extract(r'R\$ ?([\d.,]+)')
precos_carrefour1['preco'] = precos_carrefour1['preco'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)


# Extrair as duas letras após o hífen na coluna 'cidade' para criar a coluna 'uf'
precos_carrefour1['uf'] = precos_carrefour1['cidade'].str.extract(r'-\s*([A-Z]{2})')

precos_carrefour1
```

```python
# Mapeamento de estados para regiões
uf_to_region = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste', 'PE': 'Nordeste',
    'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

# Criar a nova coluna com base no mapeamento
precos_carrefour1 ['regiao'] = precos_carrefour1 ['uf'].map(uf_to_region)

precos_carrefour1 
```
